# Density attacks
### Analysis and visualization
We check if, when one image from candidate looks similar to many synthetic samples (many synthetic images cluster around a candidate) it can betray that this image was used in trainign.<br/>
We run the analysis by looking at the largest clusters of synthetic data around a candidate and checking if the candidate was from training

## Generate data for privacy threatening example

```python
# We simulate images coming from different datasets.
# An image is a vector of size 1x64

# We define synthetic as the base dataset
# Candidates from training to be a subset of slightly altered synthetic images
# Canditates from training to be a subset of altered synthetic images
# Candidates from training to be a subset of very altered synthetic images

# This is not representative of the process GAN goes throught, the goal is just to create a 
# visualization that helps with understanding the other plots through an extreme case

dim_size = 64
size = 2000

# 2000 base synthetic images as random gaussian noise
synthetic_threatening = np.random.randn(dim_size, size)

# augmenting the synthetic dataset with variation of the 2000 base images
# (additive and multiplicative noise)
synthetic_threatening = np.concatenate((synthetic_threatening, synthetic_threatening + np.random.randn(dim_size, size)*0.1, 
                                       synthetic_threatening*(0.975 + np.random.rand(dim_size, size)*0.05) + np.random.randn(dim_size, size)*0.15,
                                       np.random.randn(dim_size, size),
                                       synthetic_threatening*(0.95 + np.random.rand(dim_size, size)*0.1) + np.random.randn(dim_size, size)*0.2),
                                      axis=1)
len_syn_th = synthetic_threatening.shape[1]

# training is a randomly sampled slightly altered version of the synthetic training
train_threatening = np.random.permutation(synthetic_threatening)*(0.975 + np.random.rand(dim_size, len_syn_th)*0.05) + np.abs(np.random.randn(dim_size, len_syn_th)*0.15)
train_threatening = train_threatening[:,:1000]

# validation is a randomly sampled slightly altered version of the synthetic training
t = 0.45
val_threatening = np.random.permutation(synthetic_threatening) + (t +np.random.randn(dim_size, len_syn_th)*t)**2
val_threatening = val_threatening[:,:1000]

# testing is a randomly sampled slightly altered version of the synthetic training
t = 0.5
test_threatening = np.random.permutation(synthetic_threatening) + (t + np.random.randn(dim_size, len_syn_th)*t)**2
test_threatening = test_threatening[:,:1000]

# The candidates samples are 1000 images from each origin
candidate_th = np.concatenate((train_threatening, val_threatening, test_threatening), axis=1)

# Computing the distances between synthetic and candidates
d1 = cdist(candidate_th.T, synthetic_threatening.T)
# rank by distance
d1_sorted = np.sort(d1.ravel())
percent_2 = d1_sorted[len(d1_sorted)//50]

# Get clusters
d1_clusters = np.sum((d1 < percent_2)*1.0, axis=1)
```

## Relevant global functions

```python
def sort_by_max(a):
    '''Default sorting algorithm is ascendant and does not handle similar values well.
    This function is a quick solution to sort descending with the relevant
    handling of same values'''
    arg_list = []
    temp = np.copy(a)
    for i in range(len(a)):
        ind = np.argmax(temp)
        arg_list.append(ind)
        temp[ind] = -1
    return arg_list
```

# Supplemental: Quantitative analysis of discriminative power

## Classification power on based on cluster size

```python
def origin_pred(label_sorted, cut_off=100):
    pred = label_sorted[:cut_off]
    pred_train = pred[pred==0.0]
    
    pred_val = pred[pred==1.0]
    
    pred_test = pred[pred==2.0]
    total = len(pred_train) + len(pred_val) + len(pred_test)
    percent = [len(pred_train)/total, len(pred_val)/total, len(pred_test)/total]
    return (len(pred_train)/(len(pred_val)+len(pred_train)),
            len(pred_train)/(len(pred_test)+len(pred_train)), percent)
```

## ROC
Classifying based on cluster size (large cluster size resulting in a prediction of a training sample is considered a True positive)

```python
def roc_custom(dist_list):
    '''Compute roc curve for candidate in pixel, feature space and toy example'''
    
    # Getting labels based on origin from candidate dataset
    train_list = dist_list[:334]
    val_list = dist_list[334:667]
    test_list = dist_list[667:]

    # Running normalization on the distance to get values between 0 and 1
    # normalize val vs train
    norm_val = np.concatenate((train_list, val_list))
    norm_val = norm_val - np.min(norm_val)
    norm_val = norm_val*(1/np.max(norm_val))
    
    # normalize test vs train
    norm_test = np.concatenate((train_list, test_list))
    norm_test = norm_test - np.min(norm_test)
    norm_test = norm_test*(1/np.max(norm_test))
    
    # define positive and negative labels
    # ealier samples are always training so labeled as positive
    labels = np.concatenate((np.ones([334]), np.zeros([333])))

    # Get false positve and true positive rate
    fpr_val, tpr_val, _ = roc_curve(labels, norm_val)
    fpr_test, tpr_test, _ = roc_curve(labels, norm_test)

    return fpr_val, tpr_val, fpr_test, tpr_test
```

```python
# Get clusters
L2_sort = np.sort(L2_dist.ravel())
percent_01 = L2_sort[100000]
neighbourhood_01_L2 = np.sum((L2_dist < percent_01)*1.0, axis=1)

u_sort = np.sort(umap_dist.ravel())
percent_01 = u_sort[10000]
neighbourhood_01_umap = np.sum((umap_dist < percent_01)*1.0, axis=1)

a = np.sort(d1_clusters[:1000])[666:][::-1]
b = np.sort(d1_clusters[1000:2000])[667:][::-1]
c = np.sort(d1_clusters[2000:])[667:][::-1]
d = np.concatenate((a,b,c))
simul_clusters = d*1.0
```

```python
# Get FPR and TPR
fpr_L2_val, tpr_L2_val, fpr_L2_test, tpr_L2_test = roc_custom(neighbourhood_01_L2)
fpr_umap_val, tpr_umap_val, fpr_umap_test, tpr_umap_test =  roc_custom(neighbourhood_01_umap)
fpr_L2_val_simul, tpr_L2_val_simul, fpr_L2_test_simul, tpr_L2_test_simul = roc_custom(simul_clusters)
```
